```python
## MZD 85 xsec
# x = [1.00E-10, 1.00E-08, 1.00E-06, 1.00E-04]
x=[1.00E-05, 1.00E-04, 1.00E-03, 1.00E-02]

def func(x, a):
    vec = []
    for ii in range(len(x)):
#         vec.append(a * x[ii] + b )
        vec.append(a * np.square(x[ii]))
    
    return vec 

popt, pcov = curve_fit(func, ep, xsec85_msD5)
q=popt
# print(q)
# popt, pcov = curve_fit(func, ep2, xsec85_msD5, p0=(1.00E+07, 4.00E-12))
fitLine = func(xsec85_msD5, *popt)


# Calculate R^2
residuals = xsec85_msD5 - func(ep, *popt) # residuals time, *popt func(time, *popt)
ss_res = np.sum(residuals**2) # sum of the square residuals
ss_tot = np.sum((xsec85_msD5 - np.mean(xsec85_msD5))**2)
r_squared = 1 - (ss_res / ss_tot)


i=stats.chisquare(f_obs=xsec85_msD5, f_exp=func(x, *popt))
print(i[0])
print (i[1])
print(type(i))

#N = np.format_float_scientific(popt[0], precision=3)
#N = float(N)
#print(N)
textstr = '\n'.join((
    r'$\sigma = a \times \epsilon^{2}$', 
    r'$a = %.2E\, $' % (popt[0], ) , #Issue with format2ting scientific notation in text box, so hardcoded
#     r'$b= %.3f\, $' % (popt[1], ),
    r'$\chi^{2}/ndf = %.3f$' % (i[0]/3, ) ,))

props = dict(boxstyle='round', alpha=0.5)
plt.rcParams.update({'font.size': 13})
fig, ax = plt.subplots(figsize=(10,10))

ax.text(0.2, 0.75, textstr, transform=ax.transAxes, fontsize=20, verticalalignment='top', bbox=props)
# ax.set_ylim([0,0.02])
# ax.set_xlim([0,0.1])

ax.loglog()
ax.set_xlabel(r'$\epsilon$', fontsize=15)
ax.set_ylabel(r'$\sigma$(fb)', fontsize=15)
plt.grid(True, which="both", ls="--")
plt.plot(ep,xsec85_msD5 , 'ko', label="Data")
plt.plot(x, func(x, *popt), 'r-', label="Fitted Curve")
plt.title(r'$M_{Z_{D}}=85$ GeV', fontsize=15)
plt.legend()
fig.savefig('FitSquaredFigs/Xsec_MZD_85.pdf')
plt.show()
```
