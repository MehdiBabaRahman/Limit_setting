```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m_obs=np.array([5, 15, 20, 25, 30, 35, 36, 37, 38, 39, 40, 45, 50, 55])
Obs=np.array([3.96863, 2.82549, 3.04489, 3.3085, 2.8849, 4.07995, 3.13102, 3.03473, 6.87937, 7.57799, 7.97069, 2.89264, 3.33693, 3.97197])

m_exp=np.array([5, 15, 20, 25, 30, 35, 37, 40, 45, 50, 55])
exp=np.array([3.06069, 3.03231, 3.1143, 3.0343, 3.2164, 3.2031, 3.25783, 3.3403, 3.52974, 3.5021, 3.6234])
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(m_obs, Obs, label = 'Observed', marker = 'o',color='b')
ax.plot(m_exp, exp, label = 'Epexcted', marker='o',color='r')
plt.grid(True, which="both", ls="--")
# ax.grid(self, b=None, which='major', axis='both', **kwargs)
plt.rcParams.update({'font.size': 12})
ax.set_xlabel(r'$m_{s_{D}}$ [GeV]')
ax.set_ylabel(r'95% CL upper limit on n')
# plt.title(r'$CMS\ Preliminary\ 59.97fb^{-1} \ (13 TeV)$') 
plt.title(r'$CMS \ Preliminary$', loc='left')
plt.title(r'$59.97fb^{-1} \ (13 TeV)$', loc='right')
plt.legend(bbox_to_anchor=(1, 1))
# ax.set_facecolor('black')
plt.show()
# fig.savefig('Limits_all_samples_AnLimit_range.pdf')
```
